# 🎙️ Transcribe Audio using Whisper

```python
import whisper

# Load the small model (you can use base, medium, large for higher accuracy)
model = whisper.load_model("small")

# Use the uploaded file
filename = list(uploaded.keys())[0]

# Transcribe
result = model.transcribe(filename)

# Print transcription
print("📝 Transcription:")
print(result["text"])
```
